```python
'''Normal(Gaussian) Distribution '''
# this is one of the most important distributions.
#It is also called Gaussian distribution after the german mathematician carl friedrich Gauss.
#It fits probability distribution of many (eg:IQ scores, Heartbeat).
# we use random.normal() to get Normal Data Distribution.
# It has 3 parameters 
''' loc-(mean) where the peak of the bell exists.
     scale-(standard Deviation) how flat the graph distribution should be.
     size- The shape of the returned array. '''

from numpy import random
x=random.normal(size=(3,5))
print(x)
```

```python
x=random.normal(loc=1,scale=2,size=(2,3))
print(x)
```

```python
'''Binomial Distribution'''
# Binomial distribution is a discrete Distribution.
# it outcomes is 0 or 1 and coin(head or tail)
# it has 3 parameters
''' n=no.of trails.
    p=probability of occurence of each trail 
    size=the shape of the return
    '''


'''Descrete disribution'''
# It is a probability distribution that describes the probability of each value of a discrete random variable
```

```python
from numpy import random
x=random.binomial(n=100,p=0.2,size=10)
print(x)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate binomial distribution data
data = np.random.binomial(n=10, p=0.5, size=1000)

# Use histplot instead of distplot
sns.histplot(data, kde=False)

plt.show()
```

```python
''' visualization of Bionomial distribution '''
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the normal distribution using kdeplot
sns.kdeplot(random.normal(loc=50, scale=5, size=1000), label='normal')

# Plotting the binomial distribution using kdeplot
sns.kdeplot(random.binomial(n=100, p=0.5, size=1000), label='binomial')

plt.show()
```

```python
'''poission distribution '''
# It is a discrete distribution.
# It estimates the how many times an event can times in a specified time.
#---->parameters
# lam=rate or known no.of occurances.
# size=the shape of the returned array.

from numpy import random
x=random.poisson(lam=4, size=15)
print(x)
```

```python
'''visualization of poisson distribution'''
sns.displot(random.poisson(lam=2,size=1000))
plt.show()
```

```python
'''diff between normal and poisso'''

sns.kdeplot(random.normal(loc=50,scale=7,size=1000),label="normal")
sns.kdeplot(random.poisson(lam=50,size=1000),label="poisson")
plt.show()
```

```python
''' diff between binomial and poisson '''
sns.kdeplot(random.binomial(n=1000,p=0.01,size=1000))
sns.kdeplot(random.poisson(lam=50,size=1000))
plt.show()
```

```python
'''Uniform distribution'''
# to describes the probability where every event has equal chances of occuring.
# eg:generation of random numbers.
#-------> 3 parameters
'''
a=lower bound default=0.1
b=upper bound default=1.0
size=the shape of the returned array.
'''


x=random.uniform(size=(2,3))
print(x)
```

```python
''' Visualization of uniform distribution '''
sns.kdeplot(random.uniform(size=1000))
plt.show()
```

```python
'''Logistic Distribution'''
# logistic distribution is a continuous probablity that is similar to the normal distributon,but with heavier tails
#logistic distribution is used to desribe growth
#this one is used heavily in machine learning in logistic regression ,neural networks
#---> it has 3 parameters:
# loc=mean,where the peak is Default 0.
#scale=standard deviation ,the flatness of distribution. Default is 1.
#size= the shape of the returne darray.
#f(x;u,s):e^-(x-u)/s /s(1+e^-(x-u)/s)^2

from numpy import random
x=random.logistic(loc=1, scale=2, size=(2,3))
print(x)
```

```python
'''visualizaton of Logistic Distribution'''
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.kdeplot(random.logistic(size=1000))
plt.show()
```

```python
'''diff between logistic and normal distribution '''

sns.kdeplot(random.normal(scale=2,size=1000))
sns.kdeplot(random.logistic(size=1000))
plt.show()
```

```python
'''Multinomial Distribution'''
# it is generalization of the binomial distribution
#which admits tha only two values(success and failure).
#---> it has 3 parameters:
# n=no.of possible outcomes.
# pvals=list of probabilities of outcomes.
# size=the shape of the returned array.

x=random.multinomial(n=5,pvals=[1/5,1/5,1/5,1/5,1/5,1/5])
print(x)

# it visual representation of normal,binomial,multinomial is also same
```

```python
'''Exponential Distribution'''
# exponential distribution is used for describing time till next event 
#e.g: failure / success
#----> it has 2 paramters:
# sclae=inverse of rate (see lam in poission distribution ) default is 1.0
# size=the shape of the returned array.

x=random.exponential(scale=2,size=(2,3))
print(x)
```

```python
'''visualization of exponential distribution'''
sns.kdeplot(random.exponential(size=1000))
plt.show()

'''diff between poisson and exponential'''
# poisson distribution deals with number of occurences of an event in a time period
#exponential distribution deals with the time between these events.
```

```python
''' Chi Square Distribution'''
# it is used to verify the hypothesis.
# ------> it has 2 parameters:
# df=(degree of freedom).
# size=the shape of the returned array.

x=random.chisquare(df=2,size=(2,3))
print(x)
```

```python
'''visualization of chi sqaure '''
sns.kdeplot(random.chisquare(df=1,size=1000))
plt.show()
```

```python
'''Rayleigh Distribution'''
# it is used in signal processing.
#----> it has 2 paramters.
# scale=(standard deviation) decides how ftal the distribution will be default 1.0.
# size=the shape of the returned array.


x=random.rayleigh(scale=4,size=(2,3))
print(x)
```

```python
'''visualization '''
sns.kdeplot(random.rayleigh(size=1000))
plt.show()
```

```python
'''pareto distribution'''
# it is follows pareto's law  that is 80-20 distribution (20% factors cause 80% outcomes).
#----> it has 2 parameters:
# a=shape parameter.
# size=the shape of the returned array
x=random.pareto(a=9, size=(2,4))
print(x)
```

```python
'''pareto visualization '''
sns.histplot(random.pareto(a=2,size=1000))
plt.show()
```

```python
'''Zipf Distribution'''
# it is used to sample data based on zipf's law
'''
zipf's law:In a collection ,the nth common term is 1/n timesof the mosst commmom term.
e.g: the 5th most common word in english occurs nearly 1/5 times as often as the most common word. '''

# it has 2 parameters :
# a=distribution 
# size=the shape of the returned array.

x=random.zipf(a=2,size=(2,5))
print(x)
```

```python
'''Zipf visualization '''
x=random.zipf(a=2,size=1000)
sns.displot(x[x<10])
plt.show()
```
